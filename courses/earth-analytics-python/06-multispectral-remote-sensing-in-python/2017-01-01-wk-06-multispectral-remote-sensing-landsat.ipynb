## Plot 3: Create a Plot of the Difference NDVI Using NAIP Data from 2017 & 2015

Calculate and plot the DIFFERENCE between NDVI in 2017 and 2015. To calculate difference **subtract the pre-fire data from the post fire data (post - pre)** to ensure that negative values represent a decrease in NDVI between the two years. 

For this to work properly you will need to ensure that the 2017 data are CLIPPED to the 2015 data boundary as described for plots 1&2. 

Add the Cold Springs fire boundary to your plot. 

### Data 
Use the following data to complete this plot.

* NAIP Pre-fire:  `cold-springs-fire/naip/m_3910505_nw_13_1_20150919/crop/m_3910505_nw_13_1_20150919_crop.tif`
* NAIP Post Fire: Use the data that you got from Earth Explorer
* Fire Boundary: `data/cold-springs-fire/vector_layers/fire-boundary-geomac/co_cold_springs_20160711_2200_dd83.shp`

```python
# Calculate ndvi for 2015 and 2017 NAIP data
naip_ndvi_2015 = es.normalized_diff(naip_2015_crop[3], naip_2015_crop[0])
naip_ndvi_2017 = es.normalized_diff(naip_2017_crop[3], naip_2017_crop[0])

# Calculate NDVI Difference
ndvi_diff = naip_ndvi_2017 - naip_ndvi_2015
```

The intermediate NDVI plots below are not required for your homework. They are here so you can compare intermediate outputs if you want to! You will need to create these datasets to process the final NDVI difference plot that is a homework item!

```python
# force the z axis min an dmax to be the same
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
ndvi_2015 = ax1.imshow(naip_ndvi_2015,
                       cmap='PiYG',
                       vmin=-1, vmax=1,
                       extent=naip_2015_ext)
fire_bound_utmz13.plot(ax=ax1, color='None',
                       edgecolor='black', linewidth=2)
ep.colorbar(ndvi_2015)
ax1.set(title="NAIP Derived NDVI\n September 2015 pre-fire\n This is NOT a homework plot but here for you to compare your answer if you want.")

ndvi_2017 = ax2.imshow(naip_ndvi_2017,
                       cmap='PiYG',
                       vmin=-1, vmax=1,
                       extent=naip_2015_ext)
fire_bound_utmz13.plot(ax=ax2, color='None', edgecolor='black', linewidth=2)
ep.colorbar(ndvi_2017)
ax2.set(title="NAIP Derived NDVI\n September 2017 post-fire")

plt.tight_layout()
ax1.set_axis_off()
ax2.set_axis_off()
plt.show()
```

```python
# plot difference post fire minus pre
fig, ax = plt.subplots(figsize=(12, 6))
ndvi = ax.imshow(ndvi_diff,
                 cmap='PiYG',
                 extent=naip_2015_ext)
fire_bound_utmz13.plot(ax=ax, color='None', edgecolor='black', linewidth=2)
ep.colorbar(ndvi)
ax.set(title="Homework Plot 3: NAIP NDVI Difference -  \n Post minus Pre fire (2017 - 2015)")
ax.set_axis_off()
plt.show()
```

## Plot 4 & 5: RGB & CIR Images Using Pre-Fire Landsat Data

Plot a RGB and CIR image using Landsat data collected pre-fire.

Add the Cold Springs fire boundary to each map. 
    
### Data

Landsate Prefire Data: `cold-springs-fire/landsat_collect/LC080340322016070701T1-SC20180214145604/crop/`

```python
# Import, stack landsat data (PRE data)
all_landsat_band_paths = glob(
    "data/cold-springs-fire/landsat_collect/LC080340322016070701T1-SC20180214145604/crop/*band*.tif")
all_landsat_band_paths.sort()
landsat_pre_out = "data/cold-springs-fire/outputs/landsat_prefire.tif"

# Stack landsat tif files using es.stack_raster_tifs - earthpy
landsat_pre, landsat_pre_meta = es.stack(all_landsat_band_paths, landsat_pre_out)
extent_landsat = plotting_extent(landsat_pre[0], landsat_pre_meta["transform"])
```

```python
# HW Plots 4 & 5 go here - code to plot data only
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 11))

# RGB image
ep.plot_rgb(landsat_pre,
            rgb=[3, 2, 1],
            extent=extent_landsat,
            ax=ax1, title="Homework Plot 4: Landsat RGB Image\n Pre-Fire Data")
fire_bound_utmz13.plot(ax=ax1,
                       color='None',
                       edgecolor='white',
                       linewidth=2)
# CIR
ep.plot_rgb(landsat_pre,
            rgb=[4, 3, 2],
            extent=extent_landsat,
            ax=ax2, title="Homework Plot 5: Landsat CIR Image\n Pre-Fire Data")
fire_bound_utmz13.plot(ax=ax2,
                       color='None',
                       edgecolor='white',
                       linewidth=2)
plt.show()
```
