```python
# BEGIN SOLUTION
# Import hourly precip file
f = os.path.join("data", "colorado-flood", "precipitation",
                 "805333-precip-daily-1948-2013.csv")
precip_hourly = pd.read_csv(f,
                            parse_dates=['DATE'],
                            na_values=[999.99],
                            index_col=["DATE"])

# Create a new dataframe with resample and rid of Nan values
precip_daily = precip_hourly.resample(rule='D').sum()
precip_monthly = precip_hourly.resample(rule='M').sum()


# TO DO BELOW - test for index date time??
# END SOLUTION

# BEGIN SOLUTION

# subset the data from Jan 1 - Oct 15 2013
#precip_monthly_subset = precip_monthly["1990":"2014"]

# Create subplot and plot data
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(precip_hourly.index.values,
           precip_hourly['HPCP'],
           color='purple')

ax.set(xlabel='Date',
       ylabel='Precipitation (Inches)',
       title="HW Plot 1: Hourly Precipitation - Boulder\n 1948 - 2013")
plt.show()
# END SOLUTION
```

```python
# BEGIN SOLUTION

# create subplot and plot data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
ax1.scatter(precip_daily.index.values,
            precip_daily['HPCP'],
            color='purple')

# set titles
ax1.set(xlabel='Date',
        ylabel='Precipitation (Inches)',
        title="HW Plot 2: Daily Precipitation - Boulder\n 1948 - 2013")
ax1.set_ylim(0, 20)

ax2.scatter(precip_monthly.index.values,
            precip_monthly['HPCP'],
            color='purple')

# set titles
ax2.set(xlabel='Date',
        ylabel='Precipitation (Inches)',
        title="HW Plot 3: Monthly Precipitation - Boulder\n 1948 - 2013")
ax2.set_ylim(0, 20)

# END SOLUTION

### DO NOT REMOVE LINE BELOW ###
plt.show()
```
