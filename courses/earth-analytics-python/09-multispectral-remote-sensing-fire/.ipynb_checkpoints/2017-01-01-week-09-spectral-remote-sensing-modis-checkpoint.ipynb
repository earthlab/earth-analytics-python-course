```python
import warnings
from glob import glob
import os

import numpy as np
import numpy.ma as ma
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import rioxarray as rxr
import xarray as xr
from rasterio.plot import plotting_extent
import geopandas as gpd
from shapely.geometry import box

import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
import earthpy.mask as em

data1 = et.data.get_data('cs-test-naip')
data2 = et.data.get_data('cold-springs-fire')
data3 = et.data.get_data('cold-springs-modis-h4')
data4 = et.data.get_data(url="https://ndownloader.figshare.com/files/21941085")

os.chdir(os.path.join(et.io.HOME,
                      'earth-analytics',
                      'data'))
warnings.simplefilter('ignore')

rioxarray_option = rxr.set_options(export_grid_mapping=False)
```

```python
def open_clean_bands(band_path,
                     crop_bound,
                     valid_range=None,
                     variable=None):
    """Open and clean a single landsat band.

    Parameters
    -----------
    band_path : string 
        A path to the array to be opened.
    crop_bound : geopandas GeoDataFrame
        A geopandas dataframe to be used to crop the raster data using rioxarray clip().
    valid_range:tuple (optional)
        A tuple of min and max range of values for the data. Default = None.
    variable : List
        A list of variables to be opened from the raster.

    Returns
    -----------
    band : xarray DataArray
        Cropped xarray DataArray
    """

    crop_bound_box = [box(*crop_bound.bounds.loc[0])]

    try:
        band = rxr.open_rasterio(band_path,
                                 masked=True,
                                 variable=variable).rio.clip(crop_bound_box,
                                                             crs=crop_bound.crs,
                                                             all_touched=True,
                                                             from_disk=True).squeeze()
    except:
        raise ValueError(
            "Oops - I couldn't clip your data. This may be due to a crs error.")

    # Only mask the data to the valid range if a valid range tuple is provided
    if valid_range is not None:
        mask = ((band < valid_range[0]) | (band > valid_range[1]))
        band = band.where(~xr.where(mask, True, False))

    return band

def classify_dnbr(arr):
    """Takes an input numpy array and classifies it using standard nbr classes.

    Parameters
    ----------
    arr : xarray DataArray 
        An xarray DataArray to be classified into dnbr classes.

    Returns
    -------
    digitized : xarray DataArray
        A digitized xarray DataArray put into the proper classes.

    """

    # These are standard values so it's ok to place them in here in this particular
    # Case. If we wanted a more generic classification function then we may adjust

    dnbr_class_bins = [-np.inf, -.1, .1, .27, .66, np.inf]
    digitized = xr.apply_ufunc(np.digitize,
                               arr,
                               dnbr_class_bins)

    # Missing data are assigned a 6, mask those values from the data
    mask = (digitized > 5)
    digitized = digitized.where(~xr.where(mask, True, False))

    return digitized

def clean_array_plot(xr_obj):
    """
    Take an  xarray object and replace null  values with a mask for plotting

    Parameters
    ----------
    xr_obj : xarray object

    Returns
    -------
    A masked numpy array 

    """
    return ma.masked_array(xr_obj.values,  xr_obj.isnull())
```

```python
fire_boundary_path = os.path.join("cold-springs-fire",
                                  "vector_layers",
                                  "fire-boundary-geomac",
                                  "co_cold_springs_20160711_2200_dd83.shp")
fire_boundary = gpd.read_file(fire_boundary_path)


data_path = os.path.join("earthpy-downloads",
                         "landsat-coldsprings-hw")

# view subdirectories of data
all_dirs = glob(os.path.join(data_path, "*/"))

# High level
landsat_cleaned_data = {}

# Create loop through 1) each landsat directory and then 2) the tif files
for landsat_dir in all_dirs:

    # Get dir name - you could also chose to just use the date as the scene dictionary key
    landsat_scene = os.path.basename(os.path.normpath(landsat_dir))
    scene_date = landsat_scene[10:18]

    # Get all tif files in the directory
    all_files = sorted(glob(os.path.join(landsat_dir, "*band[3-7]*.tif")))

    # Crop and clean data
    landsat_arrays = []
    for a_band in all_files:
        landsat_arr = open_clean_bands(a_band,
                                       fire_boundary,
                                       valid_range=(0, 10000))

        landsat_arrays.append(landsat_arr)

    cleaned_landsat = xr.concat(landsat_arrays, dim="band")

    landsat_cleaned_data[scene_date] = cleaned_landsat

extent_landsat = plotting_extent(
    cleaned_landsat[0], cleaned_landsat.rio.transform())
```

```python
modis_path_dirs = glob(os.path.join("cold-springs-modis-h4", "*/"))

modis_cleaned_data = {}
desired_bands = ["sur_refl_b01_1", "sur_refl_b02_1",
                 "sur_refl_b04_1", "sur_refl_b07_1"]

for adir in modis_path_dirs:

    # Get date for data
    modis_date = os.path.basename(os.path.normpath(adir))
    modis_h4_path = glob(os.path.join(adir, "*.hdf"))[0]

    modis_cropped = open_clean_bands(modis_h4_path,
                                     fire_boundary,
                                     valid_range=(0, 10000),
                                     variable=desired_bands)

    modis_cleaned_data[modis_date] = modis_cropped.to_array()

# Reproject fire boundary
fire_bound_sin = fire_boundary.to_crs(modis_cropped.rio.crs)
extent_modis_post = plotting_extent(modis_cropped.sur_refl_b01_1.values,
                                    modis_cropped.rio.transform())
```

```python
# NAIP data
naip_path = os.path.join("cs-test-naip",
                         "m_3910505_nw_13_1_20170902.tif")

naip_data = open_clean_bands(naip_path,
                             fire_boundary)

fire_bound_utmz13 = fire_boundary.to_crs(naip_data.rio.crs)

extent_naip = plotting_extent(naip_data[0],
                              naip_data.rio.transform())
```

## Homework Figure 1 - Grid of 3 Color InfraRed (CIR) Post-Fire Plots: NAIP, Landsat and MODIS

```python
# Plot NAIP, Landsat & MODIS together

naip_plot = clean_array_plot(naip_data)
landsat_plot = clean_array_plot(landsat_cleaned_data["20160723"])
modis_plot = clean_array_plot(modis_cleaned_data["17_july_2016"])

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))
ep.plot_rgb(naip_plot,
            ax=ax1,
            rgb=[3, 2, 1],
            extent=extent_naip,
            title="Naip CIR Composite Image | 1 meter \n Post Fire")
fire_bound_utmz13.plot(ax=ax1,
                       color='None',
                       edgecolor='yellow',
                       linewidth=2)

ep.plot_rgb(landsat_plot, ax=ax2, rgb=[2, 1, 0],
            title="Landsat CIR Composite Image | 30 meters \n Post Cold Springs Fire \n July 23, 2016",
            extent=extent_landsat)
fire_bound_utmz13.plot(ax=ax2,
                       color='None',
                       edgecolor='yellow',
                       linewidth=2)

ep.plot_rgb(modis_plot,
            ax=ax3,
            rgb=[1, 0, 2],
            title="Modis CIR Composite Image | 500 meters \n Post Cold Springs Fire \n July 17, 2016",
            extent=extent_modis_post)

fire_bound_sin.plot(ax=ax3,
                    color='None',
                    edgecolor='yellow',
                    linewidth=2)

plt.show()
```

## Homework Figure 3 - Difference NBR (dNBR) Using Landsat & MODIS

```python
dnbr_cat_names = ["Enhanced Regrowth",
                  "Unburned",
                  "Low Severity",
                  "Moderate Severity",
                  "High Severity"]

nbr_colors = ["g", "yellowgreen", "peachpuff", "coral", "maroon"]
nbr_cmap = ListedColormap(nbr_colors)

landsat_veg_index = {}
for scene_date in landsat_cleaned_data:
    # Calculate veg indices
    band_3 = landsat_cleaned_data[scene_date][1]
    band_4 = landsat_cleaned_data[scene_date][2]
    band_6 = landsat_cleaned_data[scene_date][4]
    landsat_veg_index[scene_date] = {"ndvi": (band_4 - band_3)/(band_4 + band_3),
                                     "nbr": (band_4 - band_6)/(band_4 + band_6)}

# Calculate diff NDVI
landsat_ndvi_diff = landsat_veg_index['20160723']['ndvi'] - \
    landsat_veg_index['20160621']['ndvi']

# Calculate dNBR from pre-post fire nbr
dnbr_landsat_arr = landsat_veg_index['20160621']['nbr'] - \
    landsat_veg_index['20160723']['nbr']

landsat_dnbr_class = classify_dnbr(dnbr_landsat_arr)

landsat_dnbr_class
```

```python
# just here for comparison if you want to look at each plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 9))

landsat_pre = clean_array_plot(landsat_veg_index['20160621']['nbr'])
landsat_post = clean_array_plot(landsat_veg_index['20160723']['nbr'])

ep.plot_bands(landsat_veg_index['20160621']['nbr'],
              cmap='RdYlGn',
              scale=False,
              vmin=-1, vmax=1,
              ax=ax1,
              title="Landsat PRE - Normalized Burn Ratio (NBR)\n This is just a reference plot")

ep.plot_bands(landsat_veg_index['20160723']['nbr'],
              cmap='RdYlGn',
              scale=False,
              vmin=-1, vmax=1,
              ax=ax2,
              title="Landsat POST Normalized Burn Ratio (NBR)\n This is just a reference plot")
plt.show()
```

```python
modis_veg_index = {}

for modis_date in modis_cleaned_data:
    # Calculate Veg Indices
    band_3 = modis_cleaned_data[modis_date][0]
    band_4 = modis_cleaned_data[modis_date][1]
    band_6 = modis_cleaned_data[modis_date][3]
    modis_veg_index[modis_date] = {"ndvi": (band_4 - band_3)/(band_4 + band_3),
                                   "nbr": (band_4 - band_6)/(band_4 + band_6)}

# Calculate veg indices
modis_ndvi_diff = modis_veg_index["17_july_2016"]['ndvi'] - \
    modis_veg_index["07_july_2016"]['ndvi']
modis_dnbr = modis_veg_index["07_july_2016"]['nbr'] - \
    modis_veg_index["17_july_2016"]['nbr']

modis_dnbr_class = classify_dnbr(modis_dnbr)

modis_dnbr_class
```
